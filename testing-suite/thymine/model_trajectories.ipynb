```python
import os
import matplotlib.pyplot as plt
from ase.io.trajectory import Trajectory
```

```python
traj_path = 'testing_suite/thymine/mds/model_trajectories'
trajs = os.listdir(traj_path)
print(trajs)
```

```python
test_traj = 'testing_suite/thymine/mds/model_trajectories/model_mds.traj'
traj = Trajectory(test_traj)
traj[0].info
```

```python
fig, ax = plt.subplots(2,2)
ax = ax.flatten()
for i, traj in enumerate(trajs):
    traj_path = f'testing_suite/thymine/mds/model_trajectories/{traj}'
    traj_obj = Trajectory(traj_path)
    es_1_energies = [traj_obj[i].info['es1_energy'] for i in range(len(traj_obj))]
    es_2_energies = [traj_obj[i].info['es2_energy'] for i in range(len(traj_obj))]
    ax[i].plot(es_1_energies)
    ax[i].plot(es_2_energies)
    ax[i].set_title(traj)
fig.tight_layout()    
```
